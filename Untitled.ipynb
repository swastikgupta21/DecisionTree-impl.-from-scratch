```python
class DecisionTree:
    
    def _init_():
            self.left=None
            self.right=None
            self.fkey=None
            self.fval=None
            self.decision=None
            
    @staticmethod
    def entropy(col):
        total=col.shape[0]
        counts=np.unique(col,return_counts=True)
        ent=0
        
        for i in counts[1]:
            p=float(i/total)
            ent+= p*np.log2(p)
        
        ent*=-1
        return ent
    @staticmethod
    def info_gain(data_left,data_right,data):
        
        if(data_left.shape[0]==0 or data_right.shape[0]==0):
            return -100000
        
        main_entropy=DecisionTree.entropy(data)
        left_entropy=DecisionTree.entropy(data_left)
        right_entropy=DecisionTree.entropy(data_right)
        
        ig=main_entropy - float(data_left.shape[0]/data.shape[0])*left_entropy - float(data_right.shape[0]/data.shape[0])*right_entropy
        
        return ig
    @staticmethod
    def divide(data,fkey):
        
        para=np.mean(data[fkey])
        
        data_left=pd.DataFrame([],columns=data.columns)
        data_right=pd.DataFrame([],columns=data.columns)
        
        for i in range(data.shape[0]):
            
            if(data[fkey].loc[i]>para):
                data_right=data_right.append(data.loc[i])
            else:
                data_left=data_left.append(data.loc[i])
        
#         print(data_left.shape,data_right.shape)
        data_left.reset_index(drop=True,inplace=True)
        data_right.reset_index(drop=True,inplace=True)
        

        return data_left,data_right
    
    def make_dt(self,data,max_depth=5,current_depth=1):
#         print("entered")

        if(current_depth==max_depth):
            counts=np.unique(data.survived,return_counts=True)
            self.decision=counts[0][np.argmax(counts[1])]
            self.left=None
            self.right=None
            return
        
        
        n=data.shape[1]-1
        ig=[]
        
        for i in range(n):
            data_left,data_right=DecisionTree.divide(data,data.columns[i])
            ig.append(DecisionTree.info_gain(data_left,data_right,data))
        
        i=np.argmax(ig)
        data_left,data_right=DecisionTree.divide(data,data.columns[i])
#         print(data_left.shape,data_right.shape)
        
        self.fkey=data.columns[i]
        self.fval=np.mean(data[data.columns[i]])
        print("splitting on basis of "+ str(self.fkey) + " on level " + str(current_depth))
        
        if(data_left.shape[0]==0 or data_right.shape[0]==0):
            counts=np.unique(data.survived,return_counts=True)
            self.decision=counts[0][np.argmax(counts[1])]
            return 
        
        
        self.left=DecisionTree()
        self.left.make_dt(data_left,max_depth=5,current_depth=current_depth+1)
#         print("executed left")
        
        self.right=DecisionTree()
        self.right.make_dt(data_right,max_depth=5,current_depth=current_depth+1)
#         print("executed right")


    def func(self,test):
        
        
        if self.left is None or self.right is None :
            return self.decision
                
        if(test[self.fkey]>self.fval):
            return self.right.func(test)
                
                
        elif(test[self.fkey]<=self.fval):
            return self.left.func(test)
            
    
    def predict(self,test):
        
        n=test.shape[0]
        
        pred=[]
        
        for i in range(n):
            pred.append(self.func(test.loc[i]))
        
        return np.array(pred)
                
        
        

        
            
        
        
        
        
    
```
